# Model

```python
df1
```

#### Splittng the data

```python
x=df1.drop(['diagnosis'],axis=1)
y=df1['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split( x ,y, test_size=0.25, random_state=0)
```

## 1.SVM

```python
from sklearn import svm
```

```python
cls=svm.SVC(kernel='linear')
cls.fit(X_train,y_train)
```

```python
cls.score(X_train,y_train)
```

```python
cls.score(X_test,y_test)
```

## 2.Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
# create regressor object
classifier = RandomForestClassifier(n_estimators = 500, random_state = 0)
```

```python
classifier.fit(X_train, y_train)
#Train data
classifier.score(X_train,y_train)*100
```

```python
classifier.score(X_test,y_test)*100
```

## 3.Ensemble Model

### Ensemble Bagging technique

```python
from sklearn.ensemble import BaggingClassifier
bgcl = BaggingClassifier( n_estimators=50,random_state=1)
bgcl = bgcl.fit(X_train, y_train)
```

```python
bgcl.score(X_train,y_train)*100
```

```python
bgcl.score(X_test,y_test)*100
```

### Ensemble adaboost

```python
from sklearn.ensemble import AdaBoostClassifier
abcl = AdaBoostClassifier(n_estimators=100, random_state=1)
abcl = abcl.fit(X_train, y_train)
```

```python
abcl.score(X_train,y_train)*100
```

```python
abcl.score(X_test,y_test)*100
```
